  <script>
    // it is getting the element and also setting the default input to nothing
    let outputElement = document.getElementById('output');
    let currentInput = '';
    // put the value onto the output bar by adding the current input and the value of your calculation
    function appendToOutput(value) {
      currentInput += value;
      outputElement.innerText = currentInput;
    }
    function calculate() {
      // check if the evaluation will work
      try {
        currentInput = eval(currentInput).toString();
        outputElement.innerText = currentInput;
        // if there is an error then display error in the output bar
      } catch (error) {
        outputElement.innerText = 'Error';
      }
    }
    // if you click the clear button then set the input to blank
    function clearOutput() {
      currentInput = '';
      outputElement.innerText = '0';
    }
  </script>
